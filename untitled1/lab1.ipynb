```python
from matplotlib import pyplot as plt
import cv2 as cv
import numpy as np
%matplotlib inline

K = 255
x = np.arange(K + 1)
img = cv.imread('pics/gray.jpeg',0)
img_hist = cv.calcHist([img], [0], None, [256], [0,256])
```

```python
plt.rcParams["figure.figsize"] = [14, 15]
img_min = np.min(img)
img_max = np.max(img)
ret_img = img
#ret_img = np.zeros((img.shape[0],img.shape[1]),dtype = 'uint8')
for i in range(ret_img.shape[0]):
    for j in range(ret_img.shape[1]):
        ret_img[i][j] =  255*(img[i,j]-img_min)/(img_max-img_min)
    
#img_hist = cv.calcHist(img, [0], None, [256], [0,256])
#img_hist = img_hist.flatten()
stretched_hist = cv.calcHist([ret_img], [0], None, [256], [0,256])
#stretched_hist = stretched_hist.flatten()
    
plt.subplot(221)
plt.imshow(img, 'gray')
    
plt.subplot(223)
plt.plot(x, img_hist)
#plt.xlim([0,256])
    
plt.subplot(222)
plt.imshow(ret_img, 'gray')
    
plt.subplot(224)
plt.plot(x, stretched_hist)
#plt.xlim([0,256])
    
plt.savefig('pics/gray_stretch.png')
plt.show()
    
```

```python
plt.rcParams["figure.figsize"] = [14, 15]
drop=0.05
k = int(drop * img.size)
x_min, x_max = 0, 254

cnt_l, cnt_r = 0, 0
hist = img_hist.flatten()
for i in np.arange(K + 1):
    cnt_l += hist[i]
    cnt_r += hist[K-i]
    if cnt_l <= cnt_r:
        x_min+=1
        cnt_r-=hist[K-i]
        if cnt_l >= k:
            break
    else:
        x_max-=1
        cnt_l -= hist[i]
        if cnt_r >= k:
            break
print(x_min, x_max)
img_min = x_min
img_max = x_max
ret_img = img.astype('float')
for i in range(ret_img.shape[0]):
    for j in range(ret_img.shape[1]):
        ret_img[i][j] =  255*(img[i,j]-img_min)/(img_max-img_min)

ret_img = ret_img.astype('uint8')        
stretched_hist = cv.calcHist([ret_img], [0], None, [256], [0,256])
img_hist = img_hist.flatten()
stretched_hist= stretched_hist.flatten()

plt.subplot(221)
plt.imshow(img, 'gray')
    
plt.subplot(223)
plt.plot(img_hist)
plt.xlim([0,256])
    
plt.subplot(222)
plt.imshow(ret_img, 'gray')
    
plt.subplot(224)
plt.plot(stretched_hist)
plt.xlim([0,256])
    
plt.savefig('pics/gray_stretch_adv.png')
plt.show()
```

```python
plt.rcParams["figure.figsize"] = [14, 10]
res = img.copy()
# ref_img = eq_img
# ref_hist = eq_hist
ref_img = cv.imread('pics/gray2.jpg', 0)
ref_hist = cv.calcHist([ref_img], [0], None, [256], [0,256])
h_copy = img_hist.copy()
for i in range(img_hist.size):
    h_copy[i] = h_copy[i-1] + h_copy[i]
h_copy_norm = h_copy / img.size

ref_hist_copy = ref_hist.copy()
for i in range(ref_hist.size):
    ref_hist_copy[i] = ref_hist_copy[i-1] + ref_hist_copy[i]
r_h_norm = ref_hist_copy / ref_img.size

mapp = np.zeros(256)
for i in range(256):
    j=255
    while True:
        mapp[i]=j
        j=j-1
        if j<0 or h_copy_norm[i] > r_h_norm[j]:
            break
for i in range(res.shape[0]):
    for j in range(res.shape[1]):
        a = res.item(i,j)
        b = mapp[a]
        res.itemset((i,j), b)

res_hist = cv.calcHist([res], [0], None, [256], [0,256])

plt.subplot(231)
plt.imshow(img, 'gray')
    
plt.subplot(234)
plt.plot(img_hist)
plt.xlim([0,256])
    
plt.subplot(232)
plt.imshow(ref_img, 'gray')
    
plt.subplot(235)
plt.plot(ref_hist)
plt.xlim([0,256])

plt.subplot(233)
plt.imshow(res, 'gray')
    
plt.subplot(236)
plt.plot(res_hist)
plt.xlim([0,256])


    
plt.savefig('pics/gray_stretch_mtch.png')
plt.show()
```
