```python
from matplotlib import pyplot as plt
import cv2 as cv
import numpy as np
%matplotlib inline

K = 255
x = np.arange(K + 1)
image = cv.imread('pics/gray.jpeg', 0)
```

```python
def get_hist(img):
    hist_size = K + 1
    hist_range = (0, K)
    hist = cv.calcHist([img], [0], None, [hist_size], hist_range, accumulate=False)
    return hist.flatten()

def stretch(img):
    img_min = np.min(img)
    img_max = np.max(img)
    ret_img = img
    for i in range(ret_img.shape[0]):
        for j in range(ret_img.shape[1]):
            ret_img[i][j] =  255*(img[i,j]-img_min)/(img_max-img_min)
    return ret_img

def plot_stretch(img, name):
    img_hist = cv.calcHist(img, [0], None, [256], [0,256])
    stretched = stretch(img)
    #stretched = cv.cvtColor(stretched, cv.COLOR_BGR2GRAY)
    stretched_hist = cv.calcHist(stretched, [0], None, [256], [0,256])
    
    plt.subplot(221)
    plt.imshow(img, 'gray')
    
    plt.subplot(223)
    plt.imshow(img_hist)
    plt.xlim([0,256])
    
    plt.subplot(222)
    plt.imshow(stretched)
    
    plt.subplot(224)
    plt.imshow(stretched_hist)
    plt.xlim([0,256])
    
    #plt.savefig('pics/' + name + '_stretch.png')
    plt.show()
```

```python
plot_stretch(image, 'gray')
```
